```python
# ライブラリのインポート
from diffusers import AutoPipelineForText2Image
import torch

# Stable Diffusionモデルを準備
pipe = AutoPipelineForText2Image.from_pretrained("stabilityai/sd-turbo", torch_dtype=torch.float16, variant="fp16")
pipe.to("cuda")
```

```python
# プロンプト
prompt = "I would like you to design the packaging for cosmetics.Lipstick: Remove the cap and place it horizontally."
# 画像を生成
image = pipe(prompt=prompt, num_inference_steps=30, guidance_scale=5.0).images[0]
```
