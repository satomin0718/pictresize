```python
import pandas as pd
import matplotlib.pyplot as plt

# Titanic号のデータを読み込む
df = pd.read_csv("titanic.csv")

# 生存者数と死亡者数を計算する
survived = df[df["Survived"] == 1].shape[0]
not_survived = df[df["Survived"] == 0].shape[0]

# 生存者数と死亡者数の棒グラフを作成する
plt.bar(["生存者", "死亡者"], [survived, not_survived])

# グラフのタイトルを設定
plt.title("Titanic号の生存者数と死亡者数")

# グラフの軸ラベルを設定
plt.xlabel("生存状況")
plt.ylabel("人数")

# グラフの凡例を設定
plt.legend(["生存者", "死亡者"])

# グラフを表示
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Titanic data is loaded
df = pd.read_csv("titanic.csv")

# Calculate the number of survivors and dead
survived = df[df["Survived"] == 1].shape[0]
not_survived = df[df["Survived"] == 0].shape[0]

# Create a bar chart of the number of survivors and dead
plt.bar(["Survivors", "Dead"], [survived, not_survived])

# Set the title of the graph
plt.title("Titanic survivors and dead")

# Set the axis labels of the graph
plt.xlabel("Survivor status")
plt.ylabel("Number")

# Set the legend of the graph
plt.legend(["Survivors", "Dead"])

# Show the graph
plt.show()
```

```python
# Calculate the number of survivors and dead by gender
survived_male = df[df["Survived"] == 1].loc[df["Sex"] == "male"].shape[0]
survived_female = df[df["Survived"] == 1].loc[df["Sex"] == "female"].shape[0]
not_survived_male = df[df["Survived"] == 0].loc[df["Sex"] == "male"].shape[0]
not_survived_female = df[df["Survived"] == 0].loc[df["Sex"] == "female"].shape[0]

# Create a bar chart of the number of survivors and dead by gender
plt.bar(["Male survivors", "Female survivors", "Male dead", "Female dead"],
         [survived_male, survived_female, not_survived_male, not_survived_female])

# Set the title of the graph
plt.title("Titanic survivors and dead by gender")

# Set the axis labels of the graph
plt.xlabel("Gender")
plt.ylabel("Number")

# Set the legend of the graph
plt.legend(["Male survivors", "Female survivors", "Male dead", "Female dead"])

# Show the graph
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the Titanic data
df = pd.read_csv("titanic.csv")

# Calculate the number of survivors and deaths by title
survived_by_Name = df.groupby("Name")["Survived"].value_counts()

# Calculate the survival rate by title
survival_rate_by_Name = (survived_by_Name[1] / (survived_by_Name[1] + survived_by_Name[0])).to_frame()
survival_rate_by_Name.columns = ["Survival rate"]

# Set the title of the graph
plt.title("Titanic title-based survival rates")

# Set the axes labels of the graph
plt.xlabel("Name")
plt.ylabel("Survival rate")

# Plot the graph
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# タイタニック号の乗客データを読み込む
df = pd.read_csv("titanic.csv")

# 生存者を「生存」、死亡者を「死亡」としてラベル付けする
df["Survived"] = df["Survived"].apply(lambda x: "生存" if x == 1 else "死亡")

# チケットの等級ごとに生存者数を集計する
survived_by_class = df.groupby("Pclass").Survived.value_counts()

# グラフを作成する
plt.bar(survived_by_class.index, survived_by_class.values)
plt.xlabel("チケットの等級")
plt.ylabel("生存者数")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# タイタニック号の乗客データを読み込む
df = pd.read_csv("titanic.csv")

# チケット金額を適切に分割する
df["Fare"] = pd.qcut(df["Fare"], 4, labels=["安い", "普通", "高い", "とても高い"])

# チケット金額ごとに生存者数を集計する
survived_by_fare = df.groupby("Fare").Survived.value_counts()

# グラフを作成する
plt.bar(survived_by_fare.index, survived_by_fare.values)
plt.xlabel("チケット金額")
plt.ylabel("生存者数")
plt.show()
```
